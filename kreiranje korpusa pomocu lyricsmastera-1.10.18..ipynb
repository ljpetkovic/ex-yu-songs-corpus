## Funkcija za objedinjeno preuzimanje tekstova više izvođača

```python
from lyricsmaster import LyricWiki

provider = LyricWiki()
izvodjaci = ['Bijelo Dugme', 'Bajaga & Instruktori', 'Zdravko Colic']

def korpus(izvodjaci):
    try:
        for izvodjac in izvodjaci:
            discography = provider.get_lyrics(izvodjac)
            for album in discography:    # album is an Album Object.
                print('Album: ', album.title)
                for song in album:       # song is a Song Object.
                    print('Song: ', song.title)
                    print('Lyrics: ', song.lyrics)
                    discography.save()
    except AttributeError:
        print('error')
    except AttributeError:
        print('error2')

korpus(izvodjaci)
```
